```python
value = df['Region'].value_counts(normalize=True)*100
sns_plot = sns.barplot(x=value.index,y=value.values)
plt.xticks(rotation=90)

for i,v in enumerate(value.values):
    sns_plot.text(i, v, f"{v:.1f}", ha='center', va='bottom')

plt.show()
```
