# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 
jobs data.

```python
driver.get("https://www.naukri.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bangalore")
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#scraping job title from the given page
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
location_tag=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')   
for i in location_tag[0:10]:
    location=i.text
    job_location.append(location)
    
company_tag=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tag[0:10]:
    company=i.text
    company_name.append(company)
    
experience_tag=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')    
for i in experience_tag[0:10]:
    experience=i.text
    experience_required.append(experience)
    
```

```python
df=pd.DataFrame({'Title':job_title,'Loaction':job_location,'Company_name':company_name,'Experience':experience_required})
df
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
driver.get("https://www.azquotes.com/")
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
search.click()
```

```python
quote_1000=[]
author_1000=[]
type_of_quotes_1000=[]
```

```python
start = 0
end = 11

for page in range(start,end):
    quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tags:
        quote_1000.append(i.text)
    next_button=driver.find_element(By.XPATH,'//a[@rel="nofollow"]')
    next_button.click()
    time.sleep(12)
quote=(quote_1000[0:1000])


for page in range(start,end):
    author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tags:
        author_1000.append(i.text)
    next_button=driver.find_element(By.XPATH,'//a[@rel="nofollow"]')
    next_button.click()
    time.sleep(12)
author=(author_1000[0:1000])


for page in range(start,end):
    type_of_quotes_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_of_quotes_tags:
        type_of_quotes_1000.append(i.text)
    next_button=driver.find_element(By.XPATH,'//a[@rel="nofollow"]')
    next_button.click()
    time.sleep(12)
type_of_quotes=(type_of_quotes_1000[0:1000])
```

```python
df=pd.DataFrame({'Quote':quote,'Ratings':author,'Type Of Quotes':type_of_quotes})
df
```
