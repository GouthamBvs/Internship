```python
driver.get("https://archive.ics.uci.edu/")
```

```python
tab=driver.find_element(By.XPATH,'/html/body/table[1]/tbody/tr/td[2]/span[2]/a')
tab.click()
```

```python
table = driver.find_element(By.XPATH, '//table[@cellpadding="3"]')

rows = table.find_elements(By.XPATH, './/tr[position() > 1]')

name_name = []
data_type = []
default_task = []
attribute_type = []
no_of_instances = []
no_of_attributes = []
year = []


for row in rows:
    cells = row.find_elements(By.XPATH, './/td')
    if len(cells) >= 6:
        name_name.append(cells[0].text)
        data_type.append(cells[1].text.replace(" ", "-"))
        default_task.append(cells[2].text.replace(" ", "-"))
        attribute_type.append(cells[3].text.replace(" ", "-"))
        no_of_instances.append(cells[4].text.replace(" ", "-"))
        no_of_attributes.append(cells[5].text.replace(" ", "-"))
        year.append(cells[6].text.replace(" ", "-"))


df = pd.DataFrame({'Dataset name': name_name,
                   'Data type': data_type,
                   'Task': default_task,
                   'Attribute type': attribute_type,
                   'No of instances': no_of_instances,
                   'No of attributes': no_of_attributes,
                   'Year': year}, index=range(1, len(name_name)+1))
```

```python
df
```

```python
driver.get("https://www.naukri.com/hr-recruiters-consultants")
```

```python
name=[]
designation=[]
company=[]
skills_they_hire_for=[] 
location=[]

for i in range(500):
    driver.execute_script("window.scrollBy(0,1000)")

names=driver.find_elements(By.XPATH,'//div[@class="recInfo"]/div/p/a[1]/span')
time.sleep(2)
for name_txt in names:
    try:
        name.append(name_txt.text)
    except NoSuchElementException:
        name.append('-') 
        
designations=driver.find_elements(By.XPATH,'//div[@class="recInfo"]/div/p/span[1]')
time.sleep(2)
for designation_txt in designations:
    try:
        designation.append(name_txt.text)
    except NoSuchElementException:
        designation.append('-')         
                                  
companys=driver.find_elements(By.XPATH,'//div[@class="recInfo"]/div/p/a[2]/small')
time.sleep(2)
for company_txt in companys:
    try:
        company.append(company_txt.text)
    except NoSuchElementException:
        company.append('-')   
        
skills_they_hire_fors=driver.find_elements(By.XPATH,'//div[@class="recInfo"]/div[2]')
time.sleep(2)
for skills_they_hire_for_txt in skills_they_hire_fors:
    try:
        skills_they_hire_for.append(skills_they_hire_for_txt.text)
    except NoSuchElementException:
        skills_they_hire_for.append('-')
        
locations=driver.find_elements(By.XPATH,'//div[@class="recInfo"]/div/p/span[2]/small')
time.sleep(2)
for location_txt in locations:
    try:
        location.append(location_txt.text)
    except NoSuchElementException:
        location.append('-')          
```

```python
df=pd.DataFrame({'Name':name,'Designation':designation,'Company':company,'Skills they hire for':skills_they_hire_for,'Location':location})
df
```
