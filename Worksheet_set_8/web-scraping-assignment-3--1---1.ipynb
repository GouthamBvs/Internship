```python
import re

lat_lng = []
try:
    url_string = driver.current_url
    print("URL: ", url_string)
    lat_lng_str = re.findall(r'@([-0-9]+\.[0-9]+),([-0-9]+\.[0-9]+)', url_string)
    if lat_lng_str:
        lat, lng = lat_lng_str[0]
        print("Latitude: ", lat)
        print("Longitude: ", lng)
        lat_lng.append(lat)
        lat_lng.append(lng)
    else:
        print("Latitude and longitude not found in the URL.")
except NoSuchElementException:
    lat_lng.append('-')
```

```python
hostel_names=[]
hostel_dist=[]
hostel_rating=[]
hostel_review=[]
hostel_review_key=[]

hostel_name=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]/a')   
for i in hostel_name:
    hostel_names.append(i.text)
    
    
hostel_dists=driver.find_elements(By.XPATH,'//span[@class="description"]/span')   
for i in hostel_dists:
    hostel_dist.append(i.text)
    
    
hostel_ratings=driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]/div[1]')   
for i in hostel_ratings:
    hostel_rating.append(i.text)
    
hostel_reviews=driver.find_elements(By.XPATH,'//div[@class="reviews"]')   
for i in hostel_reviews:
    hostel_review.append(i.text.split(" ")[0])
    
hostel_reviews_keys=driver.find_elements(By.XPATH,'//div[@class="keyword"]')   
for i in hostel_reviews_keys:
    hostel_review_key.append(i.text)
```

```python
hostel_review_key_values=hostel_review_key[1:]
```

```python
df=pd.DataFrame({'hostel name':hostel_names,'distance from city centre':hostel_dist,'ratings':hostel_rating,'total reviews':hostel_review,'overall reviews':hostel_review_key_values})
df
```
